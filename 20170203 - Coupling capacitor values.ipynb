```python
f0 = 6E9
w0 = 2 * np.pi * f0
Cl = 214E-12 # F/m
l = 2735E-6 # m
C = Cl * l
RL = 50.
Ck = np.linspace(1E-15, 10E-15, 101)
Rstar = ( 1 + w0**2 * Ck**2 * RL**2 ) / (w0**2 * Ck**2 * RL)
QL = w0 * C * Rstar / 2.

fig = plt.figure()
common.configure_axes(13)
plt.plot(Ck*1E15, QL)
plt.xlabel("Coupling capacitance (fF)")
plt.ylabel("$Q_L$")
plt.yscale('log')
plt.grid()

common.save_figure(fig, save_path="/Users/gkoolstra/Documents/Code/iPythonNotebooks/Figures")
```

For a 0.8 $\mu$m ground plane width between the two traces, a capacitance matrix element of 23.9 pF/m was found using Q3D. To get the desired $Q_L = 10^4$, we would need 

$$ \ell = \frac{4\cdot 10^{-15}}{23.9 \cdot 10^{-12}}  = 167 \mu m$$
