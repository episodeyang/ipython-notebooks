# Derivation on of ehe driven gradient spin readout method

The basic idea of this readout is to apply a longitudinal field gradient, to create a spin dependent force on the electron.  The field gradient drives the electron periodically with a spin dependent phase (ideally 180$^\circ$ apart). The moving electron which is charged induces a voltage at the driving frequency in an pickup resonator.  The output of the pickup resonator is then compared with the drive signal in a typical microwave interferometry setup.

The problem can be expressed in terms of three harmonic oscillators.
* An input electrical oscillator which provides the gradient, described by $\omega_i, Z_i, \kappa_i$ and variable $q_i$ 
* The mechanical resonance of the trapped electron itself, described by $\omega_e, m_e, Q_e$ and variable $x$ 
* An output electrical oscillator which picks up the electron signal $\omega_o, Z_o, \kappa_o$ and variable $q_o$ 

The following derivation will assume that the input resonator drives the electron, which then drives the output resonator.  Backaction terms will be ignored.  Since these couplings are small, the backaction terms should just give very small shifts to the resonance frequencies and damping, but I think these can safely be ignored for this readout protocol. 

Questions: 
* What is the magnetic field at elevation $h$ above the surface, originating from two parallel wires
 <font color='red'>We define the directions as follows: $y$ is the direction perpendicular to the surface and $z$ is the direction in the plane. Electrons move in this direction. 
 The magnetic field in the $y,z$-direction is given by $$B_z = \frac{\mu_0 I}{2\pi} \left( \frac{h}{(z-d_1/2)^2+h^2} - \frac{h}{(z+d_1/2)^2 + h^2} \right)$$ $$B_y = \frac{\mu_0 I}{2\pi} \left( \frac{z-d_1/2}{(z-d_1/2)^2+h^2} - \frac{z+d_1/2}{(z+d_1/2)^2 + h^2} \right)$$ and consequently, the gradient is given by 
 $$\frac{\partial B_y}{\partial z} = \frac{\mu_0 I}{2\pi} \frac{-32 (d_1^3 - 12 d_1 h^2) z }{(d_1^2+4h^2)^3} $$ 
 Note that this reduces to $-32 z/d_1^3 \mu_0 I/2\pi$ when the electron is in the plane of the wires ($h = 0$). This expression for the gradient is valid for small values of $z \ll d_1$. However, it is zero for the equilibrium position of the electron $z = 0$, such that we can safely ignore this gradient. The other gradient is the one we are actually interested in: 
 $$ \frac{\partial B_z}{\partial z} = \frac{\mu_0 I}{2\pi} \frac{32 d_1 h}{(d_1^2+4h^2)^2} $$
 This gradient doesn't depend on the position of the electron and is non-zero at the equilibrium position of the electron.</font>

## Equations of motion 

$$\ddot{q_i} + \frac{\kappa_i}{2} \dot{q_i} + \omega_i^2 q_i = \omega_i^2 Q_i(t)$$

$$\delta \ddot{z} + \frac{\kappa_e}{2} \delta\dot{z} + \omega_e^2 \delta z = \frac{F_e(t)}{m_e}$$

$$\ddot{q_o} + \frac{\kappa_o}{2} \dot{q_o} + \omega_o^2 q_o = \omega_o^2 \frac{e z}{d_2} $$

The first equation describes the input oscillator, which is used to create a magnetic field gradient at the electron trap location.  A resonator is used to allow a relatively large gradient to be created with an input power compatible with the millikelvin stage of a dilution cryostat. Here $q_i$ is the charge on the input resonator, with $\omega_i$ and $\kappa_i$ being the resonant frequency and energy decay rate respectively. $Q_i(t)$ is the external input drive in units of charge.  

The second equation describes the motion of the trapped electron.  The trap frequency is given by $\omega_e$ and the decay rate by $\kappa_e$. The electron is driven by a force $F_e(t)$ which will arise from the gradient created by the drive on the first oscillator.  

The last equation describes the output resonator, which is used to detect the electron charge signal.  The resonant frequency and decay rate are given by $\omega_o$ and $\kappa_o$ respectively.  The drive on the resonator is given by the charge moving a fraction of the distance between the two plates of the oscillator.  In terms of charge this is the electron charge $e$ times its displacement, $x/d_2$ as a fraction of the distance between the plates, given by $d_2$.  

It is relatively straight-forward to calculate the force on the electron.  An electron in a magnetic field has energy
$$ U= -\vec{\mu} \cdot \vec{B} = -\mu_B \vec{B}\cdot \vec{\sigma}$$
where $\mu_B = e \hbar / 2 m_e$ is the Bohr magneton. If we assume both $\vec{B}$ and $\vec{\sigma}$ to be along the $\hat{z}$ direction, then the force felt by the electron will be
$$ F_e = - \frac{\partial U}{\partial z} = \mu_B \left( \frac{\partial B_z}{\partial z} \sigma_z + \frac{\partial B_y}{\partial z} \sigma_y \right) \approx \mu_B \left( \frac{\partial B_z}{\partial z} \sigma_z  \right)$$ 
<font color='red'> NB: $B_x$, $B_y$ are neglected here. This is resonable, since the gradient has a zero point at the electron equilibrium position.</font>
We imagine that the electron is near two wires with opposing currents.  These currents will create a gradient field which is roughly 
$$\frac{\partial B_z}{\partial z} = \frac{\mu_0 I}{2\pi} \frac{32 d_1 h}{(d_1^2+4h^2)^2}$$
<font color='red'> NB: the gradient doesn't depend on the position of the electron in the plane ($z$-coordinate). It only depends on the height of the electron and the distance $d_1$ between the two wires.</font>

Together this gives a spin dependent force
$$ F_e = \frac{\mu_0 \mu_B I}{2\pi} \frac{32 d_1 h}{(d_1^2+4h^2)^2} \sigma_z$$
where $r$ is the distance from the wires to the electron, $d_1$ is the distance between the wires, and $I$ is the current $\dot{q_i}$ from the first oscillator.

## Solving the equations of motion

Let the drive be $Q_i(t)=Q_i e^{i \omega t}$ and guess a solution of the form $A_i e^{i \omega t}$ then the first oscillator equation becomes the algebraic equation

$$-\omega^2 A_i  + i \omega A_i \frac{\kappa_i}{2} + \omega_i^2 A_i = \omega_i^2 Q_i$$
solving for $A_i$ gives
$$A_i = \frac{\omega_i^2 Q_i}{(\omega_i^2-\omega^2) + i \omega \frac{\kappa_i}{2}}$$
Now using $I(t) = \dot{q}_i(t) = i \omega A_i e^{i \omega t}$ we find a spin dependent force driving the electron motion given by
$$F_e = i \omega A_i e^{i\omega t} \frac{\mu_0 \mu_B}{2\pi} \frac{32 d_1 h}{(d_1^2+4h^2)^2} \sigma_z $$

Using this force as the drive on the electron motion we can apply the same method to solve for the motion with ansatz $z(t)=A_e e^{i \omega t}$.  

$$-A_e \omega^2+i \omega \frac{\kappa_e}{2} A_e +\omega_e^2 A_e = i \omega A_i \frac{\mu_0 \mu_B}{2\pi m_e} \frac{32 d_1 h}{(d_1^2+4h^2)^2} \sigma_z$$ 

Solving for $A_e$ gives
$$A_e = \frac{i \omega A_i}{(\omega_e^2-\omega^2)+i \omega \frac{\kappa_e}{2} } \frac{\mu_0 \mu_B}{2\pi m_e} \frac{32 d_1 h}{(d_1^2+4h^2)^2} \sigma_z$$ 

Similarly the electron motion drives creates an effective charge drive on the output resonator.  Using the same form we solve for $q_o(t) = A_o e^{i \omega t}$.

$$ -\omega^2 A_o + i \omega A_o \frac{\kappa_o}{2} + \omega_o^2 A_o = \omega_o^2 \frac{e A_e}{d_2}  $$

Solving for $A_o$ gives

$$ A_o = e \frac{A_e}{d_2} \frac{\omega_o^2}{(\omega_o^2-\omega^2) + i \omega \frac{\kappa_o}{2} }  $$

Substituting in for $A_e$ and $A_i$ we can find the amplitude of the output in terms of the drive.

$$ A_o =Q_i \left(\frac{\omega_o^2}{(\omega_o^2-\omega^2) + i \omega \frac{\kappa_o}{2} }\right) \left( \frac{i \omega}{(\omega_e^2-\omega^2)+i \omega \frac{\kappa_e}{2} }\right) \left( \frac{\omega_i^2 }{(\omega_i^2-\omega^2) + i \omega \frac{\kappa_i}{2}}\right) \frac{\mu_0 \mu_B}{2\pi m_e} \frac{32 e d_1/d_2 h}{(d_1^2+4h^2)^2} \sigma_z $$

The above expression gives us the ratio of amplitude of the charge oscillation inside the output resonator ($A_0$) with respect to the amplitude of the effective drive of the input oscillator $Q_i$.  To compare with typical experiments it is more convenient to rexpress these quantities in terms of input and output power (though one should be careful to note that the measurement is phase sensitive).  

## Discussion

Key features of the expression are the following.  Since it is just a driven harmonic system the output is a coherent tone with the same frequency as the input.  The phase (sign) of the amplitude is spin dependent ($\propto \sigma_z$) as we hoped. There are three lorentzian factors (the middle one looks a bit different because the coupling was through the derivative in this choice of variables), and a ratio of fundamental and geometry related constants which give the strength of the current $\rightarrow$ spin-dependent force $\rightarrow$ charge coupling. 

The parameters available to the experimentalist are the drive amplitude, $Q_i$, the oscillator parameters ($\omega$, $\kappa$, $Z$) for each oscillator, and the geometry ($r$, $d_1$, $d_2$). All of the geometry parameters are likely to be of similar order, and making them as small as feasible is clearly a big win $d^3$ or $d^6$ in power!  Essentially the size scale controls the coupling strength.  **Todo: what is the dimensionless ratio that this d should be compared to? ** The Lorentzian factors depend on the oscillator parameters in the usual way.  Generally one wants high Q oscillators, all in resonance with one another.  However, practically it is worth considering each one separately given there specific roles and physical origin. 

The input resonator is probably the least critical. It's primary role is to reduce the required drive power.  In principle one can always increase $Q_i$ and even eliminate this resonator.  However, practically it is convenient as it allows one to drive with much less input power, allowing large circulating currents to be developed without having much power dissipation in the cryostat.  This portion of the circuit is most appropriately thought of as an impedance matching transformer to couple power to the gradient generating wires more efficiently.  We have found that we can generate reasonable fractions of a milliamp even at high frequencies without a resonator, so even a low $Q\approx10-100$ should be more than sufficient.  Ideally the input portion of the circuit would be limited only by the RF critical current of the superconducting wires used to generate the gradient.  Making the characteristic impedance of the input resonator lower (to better match the low impedance of current generating wire) also improves the coupling somewhat.   

The output resonator serves to pick up the charge signal.  I believe this derivation implicitly assumes a large detuning or weak coupling between the electron and the output resonator.  In this case the output resonator acts enhances the radiative coupling between the electron and the output transmission line, via the Purcell effect.  In the large detuning / weak coupling limit, the higher the Q the better.  It is possible that if one can get the electron on resonance, and the electron is coherent that a modest Q is actually more effective.  We have previously estimated the charge coupling to be around $g/2\pi\approx 1-5$ MHz, so a $\kappa_o/2\pi\approx 0.1-1$ MHz should be sufficient, corresponding to $Q\approx 10^4$.  It should also be degenerate with the input resonator, and most crucially as we will see later it must have **no direct coupling** to the input resonator.  <font color='red'>Making the characteristic impedance of the output resonator higher to better match the high impedance of the charge oscillator, improves the coupling (by increasing $g$). (?) </font>

Given this rough analysis achieving appropriate parameters for the input and output coupling resonators appears quite feasible, at least in terms of quality factor and disorder.  The key unknown are the properties of the trapped electron, and its motional oscillator properties.  In principle, its frequency can be tuned continuously from $\omega_e/2\pi \approx 80$ GHz down to zero frequency.  In practice we have had trouble getting much below $15-20$ GHz before the electrons leave the trap.  We believe this can be overcome by improved trap design.  The quality factor of the electron motional oscillation is also unknown but I feel we must assume it will be low at least to start with (though we have calculated that it can be quite high in principle).  Loss mechanisms are radiative (good), non-radiative (phonons), or dephasing (helium fluctuations / charge noise). Preliminary measurements of the ensembles say that the Q may be as low as 10-100, though we hope this is mostly due to ensemble effects and will be improved for single trapped electrons.  The measured level of helium fluctuations could also give fairly large dephasing for long experiments (though hopefully can be considered static disorder for experiments of millisecond duration).

Systematics I think will be a key challenge especially in the initial identification of the spin-dependent signal.  Depending on the electron oscillator properties and how close to resonance we can tune it, we will find that the experiment can be relatively robust against systematics or extremely sensitive.  The next sections will explore the signal as a function of the system parameters, compare them to typical signals measured in qubit / micro-esr experiments, and also compare them to potential sources of leakage as well as suggest some potential mitigation strategies.  The primary challenge is that as stated the experiment there is a single frequency in the system which ideally can only pass through via a spin-dependent mechanism, however if the signal can leak from input to output the spin-dependent part may sit on a large background, requiring one to distinguish a small phase shift on top of a large signal rather than a full $180^\circ$ phase shift with random noise.

### Summary

The power inside the resonator may be expressed as $P_\mathrm{inside}$:
$$ P_\mathrm{inside} = \frac{A_i^2}{2 C} \omega_i$$
We will now look for how the internal charge $A_i$ is expressed in terms of input power (unit W). We start with the input power at the coupling port of the input resonator: 

$$ P_\mathrm{in} = 2 E_\mathrm{el} \kappa_i = \frac{A_i^2}{2 C} \kappa_i =  \frac{A_i^2}{2} Z_i \omega_i \kappa_i$$

Or conversely we can write 

$$ A_i = \sqrt{\frac{2  P_\mathrm{in}}{Z_i \omega_i \kappa_i}} $$

Done. We proceed with $A_e$. On resonance $A_e$ is written as

$$ A_e = \frac{2}{\kappa_e} \frac{\mu_B}{m_e} \frac{\mu_0}{2\pi}\frac{32 d_1 h}{(d_1^2 + 4h^2)^2} \sigma_z A_i = 
\sqrt{\frac{2  P_\mathrm{in}}{Z_i \omega_i \kappa_i}} \frac{2}{\kappa_e} \frac{\mu_B}{m_e} \frac{\mu_0}{2\pi}\frac{32 d_1 h}{(d_1^2 + 4h^2)^2} \sigma_z $$

Then the amplitude in the output resonator (again, on resonance) is 
$$A_o = \frac{e}{d_2} \frac{2\omega_0}{\kappa_0} A_e = \frac{e}{d_2} \frac{2\omega_0}{\kappa_0} \sqrt{\frac{2  P_\mathrm{in}}{Z_i \omega_i \kappa_i}} \frac{2}{\kappa_e} \frac{\mu_B}{m_e} \frac{\mu_0}{2\pi}\frac{32 d_1 h}{(d_1^2 + 4h^2)^2} \sigma_z$$

provided that $A_e < d_2$ (very important). The output power can be expressed as follows: 
$$ P_\mathrm{out} = \frac{A_o^2}{2 C} \kappa_o = \frac{A_o^2}{2} \omega_o Z_o \kappa_o = \frac{1}{2} Z_o \omega_o \kappa_o \frac{2  P_\mathrm{in}}{Z_i \omega_i \kappa_i} \left( \frac{e}{d_2} \frac{2\omega_0}{\kappa_0} \frac{2}{\kappa_e} \frac{\mu_B}{m_e} \frac{\mu_0}{2\pi}\frac{32 d_1 h}{(d_1^2 + 4h^2)^2} \sigma_z \right)^2$$

Then 

$$ \frac{P_\mathrm{out}}{P_\mathrm{in}} = \frac{Z_o \omega_o \kappa_o}{Z_i \omega_i \kappa_i} \left( \frac{e}{d_2} \frac{4\omega_o}{\kappa_e \kappa_o} \frac{\mu_B}{m_e} \frac{\mu_0}{2\pi}\frac{32 d_1 h}{(d_1^2 + 4h^2)^2} \sigma_z \right)^2 $$

Using $\mu_B = e\hbar/2m_e$ and letting $\omega_i = \omega_o = \omega_e$: 

$$ \frac{P_\mathrm{out}}{P_\mathrm{in}} = \frac{\kappa_o}{\kappa_i} \left( \frac{e}{d_2} \frac{4\omega_o}{\kappa_e \kappa_o} \frac{\mu_B}{m_e} \frac{\mu_0}{2\pi}\frac{32 d_1 h}{(d_1^2 + 4h^2)^2} \right)^2 $$

Is the following statement then correct?
$$\hbar g = \mu_B \left.\frac{\partial B}{\partial z}\right|_0 z_0 \sigma_z $$

### Piece by piece

driving $Q_i \rightarrow A_i \rightarrow A_e \rightarrow A_o$ output signal

#### 1. Input resonator amplitude $A_i/Q_i$

$A_i$ describes the amplitude inside the input resonator. $Q_i$ is the drive. Note that the amplitude inside the resonator can indeed be bigger than 1. In fact, on resonance the amplitude is roughly the Q factor as one would expect. 

```python
omega = 2*pi*linspace(4.95e9, 5.05e9, 1E4)
wi = 2*pi*5e9 # resonance frequency of the input resonator.
Qi = 1E3 # Q factor of the input resonator, NOT the input charge drive.
ki = wi/Qi
ai_qi = wi**2./ ((wi**2 - omega**2) + 1j*omega*ki/2.)

print "Qi = %.0e"%(wi/ki)

fig = plt.figure(figsize =(6.,4.))
common.configure_axes(15)
plt.xlim(np.min(omega)/(2*pi*1E9), np.max(omega)/(2*pi*1E9));
plt.plot(omega/(2*pi*1E9), np.abs(ai_qi), lw=2.0)
plt.yscale('log')
plt.ylabel("Amplitude inside input resonator $A_i/Q_i$")
plt.xlabel('Probe frequency $\omega/2\pi$ (GHz)')
```

#### 2. Electron motion $A_e$

We set the current to $I =\omega A_i$, and since we will keep $2\pi$ in the definitions of $\omega$, the definition of the variable Current has a $2\pi$. In the code below, <pre>Current</pre> is the current when the probe frequency is resonant with the electron motion, i.e. it is actually defined as 

$$ I = \omega_e A_i $$

Off resonant, the current scale

```python
from mpltools import color

we = 2*pi*5E9
Qe = 1E3
ke = we/Qe
omega = 2*pi*linspace(4.95e9, 5.05e9, 1E4) #2*pi*5e9

muB = 9.27E-24 
me = 9.11E-31
dB_dz_I = 1E6 # Magnetic field gradient per Ampere in the input resonator.

fig = plt.figure(figsize =(7.,4.))
common.configure_axes(15)

for Current in 2*pi*np.array([1E-5, 1E-4, 1E-3, 1E-2]):
    Ae = 1j*Current/(we**2-omega**2 + 1j*ke*omega/2.) * muB/me * dB_dz_I
    plt.plot(omega/(2*pi*1E9), np.abs(Ae), lw=2.0, label='I = %.2f mA'%(Current/(2*pi)*1E3))
    plt.xlim(np.min(omega)/(2*pi*1E9), np.max(omega)/(2*pi*1E9));

plt.yscale('log')
plt.ylabel("Electron displacement amplitude $A_e$ (m)")
plt.xlabel('Probe frequency $\omega/2\pi$ (GHz)')
plt.grid()
common.legend_outside()

common.save_figure(fig)

I = 2*np.pi*logspace(-5, -3, 1E3)

fig2 = plt.figure(figsize =(6.,4.))
common.configure_axes(15)
color.cycle_cmap(3, cmap=plt.cm.spectral)
for Qe in (1E2, 1E3, 1E4):
    ke = we/Qe
    Ae = 1j*I/(1j*ke*we/2.) * muB/me * dB_dz_I
    plt.plot(I/(2*pi), np.abs(Ae), lw=2.0, label='$Q_e=%g$'%Qe)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Circulating current (A)')
plt.ylabel('Electron $z$ amplitude on resonance (m)')
plt.grid()
plt.legend(loc=2)

common.save_figure(fig2)
```

#### 3. Output amplitude $A_o$

$A_o$ is the charge amplitude inside the output resonator. We can get the voltage as follows
$$ V_o = A_o / C$$ and we can now estimate the electrical power inside the resonator. For simplicity, let us assume that the total energy is twice the energy of the electrical energy (which is guaranteed on resonance). $$ E_o = 2 E_{o,e} = \frac{1}{2} C V_o^2 = \frac{A_o^2}{2 C} $$
When we know the power inside the resonator, we may compute the output power as follows: 
$$P_\mathrm{out} = \frac{A_o^2}{2 C} \kappa_o$$
where $\kappa_o$ is the $\omega_o/Q_o$ is the linewidth of the output resonator.

```python
we = 2*pi*linspace(4e9, 6e9, 1E5)
omega = 2*pi*linspace(4.95e9, 5.05e9, 1E4) 
wo = 2*pi*5e9 # frequency of the tuning fork resonator
Qo = 1E4 # Q factor of the tuning fork resonator
e = 1.602E-19 # electron charge
d2 = 1E-6 # distance between two prongs of the tuning fork resonator
Displacement = (1E-6, -1E-6) # amplitude of the electron motion. Cannot be bigger than d2
Z0 = 50. # effective impedance of the resonator
Capacitance = 1/(wo*Z0) # effective capacitance of the resonator 


ko = wo/Qo

fig = plt.figure(figsize =(12.,4.))
common.configure_axes(15)

for k, z in zip(['-', '+'], Displacement): 
    Ao = e*z/d2 * wo**2./ ((wo**2 - omega**2) + 1j*omega*ko/2.)
    phase = common.get_phase(Ao)
    
    plt.subplot(121)
    plt.title('Output magnitude')
    plt.plot(omega/(2*pi*1E9), 10*log10(0.5*(np.abs(Ao)**2/Capacitance)*ko) + 30, lw=2.0, label='$\sigma_z = %s1$'%k)
    #plt.yscale('log')
    plt.xlabel('$\omega/2\pi$ (GHz)')
    plt.ylabel('Output power $P_o (\omega)$ (dBm)')
    plt.xlim(np.min(omega)/(2*pi*1E9), np.max(omega)/(2*pi*1E9));
    plt.legend()
    
    plt.subplot(122)
    plt.title('Output phase')
    plt.plot(omega/(2*pi*1E9), unwrap(phase)/pi, lw=2.0, label='$\sigma_z = %s1$'%k)
    plt.xlabel('$\omega/2\pi$ (GHz)')
    plt.ylabel('Output phase ($\pi$ rad)')
    plt.xlim(np.min(omega)/(2*pi*1E9), np.max(omega)/(2*pi*1E9));
    plt.legend()
    
#print "Qi = %.0e"%(wo/ko)
    
fig = plt.figure(figsize =(6.,4.))
common.configure_axes(15)
plt.plot(omega/(2*pi*1E9), np.abs(Ao)/Capacitance, lw=2.0)
plt.yscale('log')
plt.xlabel('$\omega/2\pi$ (GHz)')
plt.ylabel('Voltage inside the resonator $V_o (\omega)$')
plt.xlim(np.min(omega)/(2*pi*1E9), np.max(omega)/(2*pi*1E9));
```
